# NumPy demo

```python
import numpy as np
from matplotlib import pyplot as plt
```

## Array generation and basic vectorized routines

```python
x = np.linspace(0., 2*np.pi, 100)
y = np.sin(x)
```

```python
%matplotlib inline
plt.plot(x, y, 'r-')
```

## NumPy routines tend to be significantly faster

```python
arr = np.arange(100000)
%timeit y = np.sin(arr)
```

```python
from math import sin

l = range(100000)
%timeit [sin(elem) for elem in l]
```

## NumPy supports Various Forms of Dimension Munging

```python
var = np.arange(10000)
var.shape
```

```python
square = var.reshape((100, 100))
square.shape
```

```python
flat = square.flatten()
flat.shape
```

```python
alsoflat = np.arange(10000)
print(np.hstack((flat, alsoflat)).shape)
print(np.vstack((flat, alsoflat)).shape)
print(np.dstack((flat, alsoflat)).shape)
```

```python
plt.imshow(square)
```

```python
plt.imshow(np.sin(square))
```

## Some mathematical functions

```python
a = np.array([0.0, 2.0, 4.0])
b = np.array([0.0, 3.0, 6.0])
```

```python
# default product is element-wise (hadamard)
a * b
```

```python
# vector scalar multiplication scales (as you'd expect)
a * 3.3
```

```python
# vector scalar addition, subtraction etc. broadcasts
1 - a
```

```python
# dot product for vectors is fairly straight forward
np.dot(a, b)
```

```python
# can get matrix multiply / dot product for 2d arrays as well
```

```python
A = np.arange(6).reshape((2, 3))
B = np.arange(6).reshape((3, 2))

print(A)
print()
print(B)

np.dot(A, B)
```

```python
np.matmul(A, B)
```

```python
np.outer(A, B)
```
