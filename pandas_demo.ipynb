## Data transformations and aggregate query patterns

```python
import re

def float_dollars(dollarstr):
    """Returns first numeric value after $ currency indicator."""
    matched = re.search('(?<=\$)\d+\.\d+', dollarstr)
    return float(matched.group(0))
```

```python
kidney[" Average Covered Charges "].apply(float_dollars).groupby(kidney["Provider State"]).mean()
```

```python
kidney[" Average Covered Charges "].apply(float_dollars).hist(bins=25)
```
