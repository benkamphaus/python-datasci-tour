```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("loan.csv", index_col="id")
```

```python
df.head()
```

```python
df.describe()
```

```python
df.columns
```

```python
label = (df["loan_status"] == "Fully Paid") | (df["loan_status"] == "Current")
```

```python
np.mean(label)
```

```python
quant_feats = ["loan_amnt", "int_rate", "dti", "annual_inc"]
```

```python
cat_feats = ["grade", "home_ownership"]
```

```python
df[quant_feats + cat_feats]
```

```python
pd.value_counts(df["grade"])
```

```python
from sklearn import preprocessing

enc = preprocessing.LabelEncoder()
enc.fit(df["grade"])
grade_nums = enc.transform(df["grade"])
grade_nums
```

```python
enc.inverse_transform(grade_nums)
```

```python
pd.value_counts(df["home_ownership"])
```

```python
grade_dummies = pd.get_dummies(df["grade"])
ho_dummies = pd.get_dummies(df["home_ownership"])
feat_df = df[quant_feats].join(grade_dummies).join(ho_dummies)
feat_df.head()
```

```python
feat_df = feat_df.fillna(0.)
```

```python
print(len(label), len(feat_df))
```

```python
from sklearn import cross_validation as cv

X_train, X_test, y_train, y_test = cv.train_test_split(feat_df, label, stratify=label)
```

```python
len(y_train), len(y_test), np.mean(y_train), np.mean(y_test)
```

```python
X_train
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
```

```python
skf = cv.StratifiedKFold(y_train, n_folds=3)

for in_idx, out_idx in skf:
    X_in, y_in = X_train.iloc[in_idx], y_train.iloc[in_idx]
    X_out, y_out = X_train.iloc[out_idx], y_train.iloc[out_idx]
    
    dtc = DecisionTreeClassifier(max_depth=20)
    dtc.fit(X_in, y_in)
    y_out_hat = dtc.predict(X_out)
    y_in_hat = dtc.predict(X_in)
    
    acc = metrics.accuracy_score(y_out, y_out_hat)
    roc_auc = metrics.roc_auc_score(y_out, y_out_hat)
    tr_acc = metrics.accuracy_score(y_in, y_in_hat)
    tr_roc_auc = metrics.roc_auc_score(y_out, y_out_hat)
    print("Training Accuracy", tr_acc)
    print("Training ROC AUC", tr_roc_auc)
    print("Out of Fold Accuracy", acc)
    print("Out of Fold ROC AUC", roc_auc)
```

```python
from sklearn.grid_search import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier()
params = {"max_depth": (5, 7, 10),
         "n_estimators": (10, 20)}
est = GridSearchCV(rfc, params)
est.fit(X_train, y_train)
```

```python
y_tr_hat = est.predict(X_train)
metrics.roc_auc_score(y_tr_hat, y_train)
```

```python
y_test_hat = est.predict(X_test)
metrics.roc_auc_score(y_test_hat, y_test)
```
